```python
import tensorflow as tf
import cv2 as cv
from matplotlib import pyplot as plt
import digitrecog
```

```python
(_, _), (x_test, _) = tf.keras.datasets.mnist.load_data()
```

```python
x_test = x_test.astype("float32") / 255
```

```python
recognizer = digitrecog.numReader()
```

```python
i=90

img= x_test[i]
print(recognizer.predict(img))
plt.imshow(img)
plt.show()

resized_img = cv.resize(img, dsize=(35,50))
print(recognizer.predict(resized_img))
plt.imshow(resized_img)
plt.show()

resized_img = cv.resize(img, dsize=(70,40))
print(recognizer.predict(resized_img))
plt.imshow(resized_img)
plt.show()
```
