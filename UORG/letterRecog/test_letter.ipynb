```python
letterRec = letterRecog.letterReader()
```

```python
from emnist import extract_test_samples
x_test, _ = extract_test_samples('letters')

x_test = x_test.astype("float32") / 255
```

```python
i=89

img= x_test[i]
print(letterRec.predict(img))
plt.imshow(img)
plt.show()

resized_img = cv.resize(img, dsize=(35,50))
print(letterRec.predict(resized_img))
plt.imshow(resized_img)
plt.show()

resized_img = cv.resize(img, dsize=(70,40))
print(letterRec.predict(resized_img))
plt.imshow(resized_img)
plt.show()
```
